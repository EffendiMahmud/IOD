# Numpy Exploration

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

# get matplot to show charts inline
%matplotlib inline
import seaborn as sns
sns.set()
````

````python
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

**1. Find Dimension of Array**

Find dimension of each array: height, weight, population
```
# Expected Ouput

Dimension(weight): 1

```

````python
print("Dimension(height): ", height.ndim)
print("Dimension(weight): ", weight.ndim)
print("Dimension(population): ", population.ndim)
````

**8. Create a histogram of height and weight**

````python
#help(plt.hist)
plt.hist(height,bins = 20, alpha=0.5)
plt.xlabel("Height (in meters)")
plt.ylabel("Count")
plt.show
````

````python
fig, ax = plt.subplots(figsize=(10,5))
ax.hist(weight)
ax.set(title='Sample normal distribution of weight', xlabel='Weight in kg', ylabel='Count')
plt.show()
````

````python
plt.subplots(nrows=1, ncols=2)
````

````python
fig,(ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.hist(weight)
ax2.hist(height)
````

````python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.hist(weight)
ax1.set_title('Weight')
ax2.hist(height)
ax2.set_title('Height')
plt.show()
````

````python
# Create a figure and an axis
fig, ax = plt.subplots(figsize = (20,10))
#
ax.hist(weight)

# Specify that we will be using a twin x axis
ax2 = ax.twinx()

ax2.hist(height*100, color = 'blue', alpha = 0.5)
plt.show()
````

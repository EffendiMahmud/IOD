# 5. Visualization



#### 5.A Plot Histogram of Age 

````python
# A:
plt.figure(figsize=(10,5))
plt.hist(df['age'], bins = 45, alpha=0.5, color='magenta')
plt.xlabel("Age", fontsize =25, color='blue')
plt.ylabel("Count", fontsize =25, color='blue')
plt.show
````

#### 5.B BoxPlot of Age

````python
# A:
plt.boxplot(df['age'])
plt.xlabel("Age")
plt.ylabel("Count")
plt.show
````

#### 5.C Scatterplot of age vs. length_of_service

````python
# A:
plt.scatter(df['age'],df['length_of_service'], alpha=0.5, color='grey')
plt.xlabel("Age")
plt.ylabel("Length of Service")
````

#### 5.D Plot Histogram of store_name

````python
# A:
plt.figure(figsize=(10,5))
plt.hist(df['store_name'], bins = 45, alpha=0.5, color='magenta')
plt.xlabel("Age", fontsize =25, color='blue')
plt.ylabel("Count", fontsize =25, color='blue')
plt.show
````

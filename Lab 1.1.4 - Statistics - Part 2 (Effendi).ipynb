````python
# calculate statistical information of a set of observations using numpy and stats
obs = np.array([21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10])

# mean
print("mean of observations = " + str(np.mean(obs)))

# median
print("median of observations = " + str(np.median(obs)))

# mode
modeResult = stats.mode(obs)
print("mode of observations = " + str(modeResult))

# percentile
#from statistics import quantiles
Q = quantiles(obs, n=4)
print("percentile of observations = " + str(Q))
Q = quantiles(obs, n=2)
print("percentile of observations = " + str(Q))
Q = quantiles(obs, n=3)
print("percentile of observations = " + str(Q))
#print("percentile of observations = " + str(np.percentile(obs, 50, axis =0, keepdims=True, out = out)))

# statistical summary of observations
print("statistical summary of observations = " + str(stats.describe(obs)))
````

# 7. [BONUS] Plot X, Y and Best Fit Line

**Write a function to calculate best fit.**

![alt text](https://i.imgur.com/B43vOV0.jpg)[1]

[1]: https://www.mathsisfun.com/data/least-squares-regression.html "Least Squares Regression"

````python
from statistics import mean
# sample points 
X = np.array([0, 5, 10, 15, 20])
Y = np.array([0, 7, 10, 13, 20])

# solve for m and b
def best_fit(X, Y):
   
    m = 1
    b = 1
    m = (((mean(X)*mean(Y)) - mean(X*Y)) /
         ((mean(X)*mean(X)) - mean(X*X)))
    
    b = mean(Y) - m*mean(X)
    
    return m, b


# solution
m, b = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line
plt.plot(X, Y, 'o')
plt.plot(X, m*X + b)
````

````python
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    # Calculate m, h
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h

data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
print(mean_confidence_interval(data))
````
